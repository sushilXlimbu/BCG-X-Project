## 3. Feature engineering

### Difference between off-peak prices in December and preceding January

Below is the code created by your colleague to calculate the feature described above. Use this code to re-create this feature and then think about ways to build on this feature to create features with a higher predictive power.

```python
price_df = pd.read_csv('..\price_data.csv')
price_df["price_date"] = pd.to_datetime(price_df["price_date"], format='%d/%m/%Y')
price_df.head()
```

```python
# Group off-peak prices by companies and month
monthly_price_by_id = price_df.groupby(['id', 'price_date']).agg({'price_off_peak_var': 'mean', 'price_off_peak_fix': 'mean'}).reset_index()
# Get january and december prices
jan_prices = monthly_price_by_id.groupby('id').first().reset_index()
dec_prices = monthly_price_by_id.groupby('id').last().reset_index()
dec_prices.head()
```

```python
# Calculate the difference
diff = pd.merge(dec_prices.rename(columns={'price_off_peak_var': 'dec_1', 'price_off_peak_fix': 'dec_2'}), jan_prices.drop(columns='price_date'), on='id')
diff.head()
```

```python
diff['offpeak_diff_dec_january_energy'] = diff['dec_1'] - diff['price_off_peak_var'] #adding new columns
diff['offpeak_diff_dec_january_power'] = diff['dec_2'] - diff['price_off_peak_fix']
diff = diff[['id', 'offpeak_diff_dec_january_energy','offpeak_diff_dec_january_power']] #only selecting required columns
diff.head()
```

```python
date_cols = ['date_activ', 'date_end', 'date_modif_prod', 'date_renewal']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# New features from dates
df['contract_length_days'] = (df['date_end'] - df['date_activ']).dt.days
df['days_since_modif'] = (df['date_renewal'] - df['date_modif_prod']).dt.days
df['months_active'] = (df['date_end'] - df['date_activ']).dt.days // 30
df.head()
```

## Merge Price Data

```python
price_df['price_date'] = pd.to_datetime(price_df['price_date'], errors='coerce')
df = df.merge(price_df, on='id', how='left')
```

```python
df['is_high_power'] = (df['pow_max'] > df['pow_max'].median()).astype(int)
df['multi_product'] = (df['nb_prod_act'] > 1).astype(int)
df.head()
```
