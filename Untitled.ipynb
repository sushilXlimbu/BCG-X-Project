```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
client_df = pd.read_csv("client_data.csv")
price_df = pd.read_csv("price_data.csv")

```

```python
date_cols = ['date_activ', 'date_end', 'date_modif_prod', 'date_renewal']
for col in date_cols:
    client_df[col] = pd.to_datetime(client_df[col], errors='coerce',dayfirst=True)
```

```python
client_df.hist(figsize=(15, 12), bins=30)
plt.tight_layout()
plt.show()
```

```python
churn = client_df[['id', 'churn']]
churn.columns = ['Client', 'churn']
churn_total = churn.groupby(churn['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
churn_percentage.columns = [ 'Percentage']
churn_percentage = churn_percentage.reset_index()

churn_percentage['churn'] = churn_percentage['churn'].replace({0: 'Retention', 1: 'Churn'})

plt.bar(churn_percentage['churn'],churn_percentage['Percentage'],color=['green','red'])
plt.ylabel('Percentage')
plt.title('Churn Distribution')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

The chart paints a largely positive picture. The tall green bar, representing Retention, shows that approximately 90% of customers have remained loyal to the company. This indicates a strong and healthy customer base.
However, the smaller red bar, representing Churn, reveals that about 10% of customers have left. While this is a minority, it represents a significant opportunity for improvement. This visual breakdown makes it immediately clear that while the company is succeeding in keeping most of its customers, there is a segment that needs attention to understand their reasons for leaving.

```python
client_df['contract_length_days'] = (client_df['date_end'] - client_df['date_activ']).dt.days
```

```python
client_df.head()
```

```python
sns.boxplot(data=client_df, x='churn', y='contract_length_days')
```

The key takeaway from this boxplot is that customers with longer contract durations are less likely to churn. The median and overall distribution of contract_length_days are noticeably higher for customers who remained with the service (churn = 0) compared to those who left (churn = 1). This suggests that contract length is a meaningful factor in customer retention and could be a useful predictor in a churn model.

```python
merged_df = pd.merge(client_df, price_df, on='id', how='left')
merged_df.head()
```

```python
numeric_df = merged_df.select_dtypes(include=['float64', 'int64'])
numeric_df.head()
```

```python
plt.figure(figsize=(15, 8))  # width=12, height=8 (in inches)
sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f", cmap='coolwarm', center=0)
plt.show()
```
